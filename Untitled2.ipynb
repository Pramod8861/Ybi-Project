# Women Cloth Reviews Prediction with Multi Nomial Naïve Bayes
---
Multinomial Naïve Bayes algorithm is a probabilistic learning method that is mostly used in Natural Language Processing. It is also suitable for text classification with discrete feature. In this project it will be used to build a women cloth review prediction model.

### Import Libary

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import Dataset

```python

df = pd.read_csv("https://raw.githubusercontent.com/YBIFoundation/ProjectHub-MachineLearning/main/Women%20Clothing%20E-Commerce%20Review.csv")

```

### Data Visualization


# **Missing Values**

```python
df.isna().sum()
```

```python
df[df['Review']==""] = np.NaN
```

```python
df['Review'].fillna("No review is given", inplace=True)
```

```python
df.isna().sum()
```

```python
df['Review']
```

## **Define Target Variable (y) and Feature Variables (X) bold text**

```python
df.columns
```

```python
x = df['Review']
y = df['Rating']
df['Rating'].value_counts()
```

## **Train Test Split**

# Getting Tokens from Feature Text

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, stratify=y, random_state=2529)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(lowercase=True, analyzer='word', ngram_range=(2, 3), stop_words='english', max_features=50000)
x_train = cv.fit_transform(x_train)
cv.get_feature_names_out()
```

```python
x_train.toarray()
```

```python
x_test = cv.fit_transform(x_test)
cv.get_feature_names_out()
```

```python
x_test.toarray()
```

## Model training

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train, y_train)
```

## Model prediction

```python
y_pred = model.predict(x_test)
y_pred.shape
```

```python
y_pred
```

##Getting probability of each predicted class

```python
model.predict_proba(x_test)
```

### Model evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```

### Recategorizing ratings as Poor (0) and Good (1)

```python
df["Rating"].value_counts()
```

```python
df.replace({'Rating': { 1:0, 2:0, 3:0, 4:1, 5:1 }}, inplace=True)
y = df['Rating']
x = df['Review']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.7, stratify=y, random_state=2529)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

### Getting Tokens from Feature Text

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(lowercase=True, analyzer='word', ngram_range=(2, 3), stop_words='english', max_features=50000)
x_train = cv.fit_transform(x_train)
x_test = cv.fit_transform(x_test)
```

### Model re-training

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_train, y_train)
```

### Model prediction

```python
y_pred = model.predict(x_test)
y_pred.shape
```

```python
y_pred
```

### Model evaluation

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))
```

```python
print(classification_report(y_test, y_pred))
```
